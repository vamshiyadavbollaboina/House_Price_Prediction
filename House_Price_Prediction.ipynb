# Importing Libraries

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# Importing Dataset

```python
data=k=pd.read_csv("/content/Housing.csv")
print(data.head())
```

# Describing dataset

```python
print(data.describe())
```

# Checking for missing values

```python
data.isnull().sum
```

# Seperating numerical and categorical data

```python
numeric_data=data.select_dtypes(include=[np.number])
categorical_data=data.select_dtypes(exclude=[np.number])
print(numeric_data)
print(categorical_data)
```

# Encoding categorical data using LabelEncoder

```python
from sklearn.preprocessing import LabelEncoder
cate_col=["mainroad","guestroom","basement","hotwaterheating","airconditioning","prefarea","furnishingstatus"]
label=LabelEncoder()
data[cate_col]=data[cate_col].apply(label.fit_transform)
```

# Visualizing histogram for numerical columns

```python
plt.figure(figsize=(8,6))
for column in numeric_data:
  sns.histplot(data=data,x=column,kde=True)
  plt.show()
```

# Dependent and Independent variable

```python
x=data.iloc[:,:-1].values
y=data.iloc[:,-1].values
```

# Train test and split

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

# Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
```

# Evaluating model

```python
print("intercept",lr.intercept_)
print(lr.coef_)
```

# Predicting test result

```python
pred=lr.predict(x_test)
print(pred)
```

# Visualizing the distribution of difference between actual and predicted values

```python
sns.distplot((y_test-pred),bins=50)
```

# Evaluation Metrics

```python
from sklearn.metrics import mean_squared_error,r2_score,mean_absolute_error
mse=mean_squared_error(y_test,pred)
sqr=np.sqrt(mse)
rs=r2_score(y_test,pred)
mae=mean_absolute_error(y_test,pred)
print("Root mean squared error : ",sqr)
print("R_squared score : ",rs)
print("mean absolute error : ",mae)
```

# Determining correlation

```python
correlation=data.corr()
plt.figure(figsize=(10,10))
sns.heatmap(correlation,cbar=True,square=True,fmt='.2f',annot=True,annot_kws={'size':8},cmap='Blues')
```
